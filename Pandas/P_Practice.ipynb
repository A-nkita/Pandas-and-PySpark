```python
import pandas as pd
```

```python
df = pd.read_csv('Amazon.csv', low_memory=False)
```

```python
df.head()
```

```python
# List of columns to retain
columns_to_keep = ['Order ID', 'Status', 'Category', 'Size', 'Qty', 'ship-state']

# Rename columns for clarity
columns_rename = {'Order ID': 'Order', 'ship-state': 'State'}

# Retain only the specified columns
df = df[columns_to_keep]

# Rename the columns
df.rename(columns=columns_rename, inplace=True)

# Save to a new CSV file
df.to_csv('order.csv', index=False)

print("New CSV file 'order.csv' created successfully.")
```

```python
# Keep only the first 20 rows
df = df.iloc[:20]

# Save to a new CSV file
df.to_csv('order.csv', index=False)

print("New CSV file 'order.csv' with first 20 rows created successfully.")
```
