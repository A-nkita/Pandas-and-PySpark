```python
#1. Create a DataFrame from a dictionary and display it.
print("Answer #1.")
import pandas as pd
data = {'Name': ['Alice', 'Bob', 'Charlie'], 'Age': [25, 30, 35], 'City': ['NY', 'LA', 'SF']}
df = pd.DataFrame(data)
print(df)

#2. Read a CSV file into a DataFrame and display its first 5 rows.
print("Answer #2.")
# Create a sample CSV file for demonstration
df.to_csv('data.csv', index=False)
df = pd.read_csv('data.csv')
print(df.head())

#3. Add a new column to a DataFrame and display the updated DataFrame.
print("Answer #3.")
df['Salary'] = [50000, 60000, 70000]
# Ensure it matches the length of the DataFrame
while len(df['Salary']) < len(df):
    df['Salary'] = df['Salary'].append(pd.Series([0]))
print(df)

#4. Filter rows based on a condition and display the result.
print("Answer #4.")
filtered_df = df[df['Age'] > 30]
print(filtered_df)

#5. Group data by a column and calculate the mean for another column.
print("Answer #5.")
grouped = df.groupby('City')['Age'].mean()
print(grouped)
```

```python
#6. Sort the DataFrame by a column in descending order.
print("Answer #6.")
sorted_df = df.sort_values(by='Age', ascending=False)
print(sorted_df)

#7. Drop a column from a DataFrame and display the result.
print("Answer #7.")
df_dropped = df.drop(columns=['Salary'])
print(df_dropped)

#8. Find the unique values in a column.
print("Answer #8.")
unique_cities = df['City'].unique()
print(unique_cities)

#9. Replace a value in a column and display the result.
print("Answer #9.")
df_replaced = df.replace({'City': {'NY': 'New York'}})
print(df_replaced)

#10. Check for missing values in the DataFrame.
print("Answer #10.")
print(df.isnull().sum())
```

```python
#11. Fill missing values in a column with a default value.
print("Answer #11.")
df['Salary'].fillna(0, inplace=True)
print(df)

#12. Rename a column in the DataFrame.
print("Answer #12.")
df.rename(columns={'Name': 'Full Name'}, inplace=True)
print(df)

#13. Reset the index of a DataFrame.
print("Answer #13.")
df_reset = df.reset_index(drop=True)
print(df_reset)

#14. Set a column as the index of the DataFrame.
print("Answer #14.")
df_indexed = df.set_index('Full Name')
print(df_indexed)

#15. Create a pivot table from the DataFrame.
print("Answer #15.")
pivot = df.pivot_table(index='City', values='Salary', aggfunc='sum')
print(pivot)
```

```python
#16. Concatenate two DataFrames vertically.
print("Answer #16.")
df1 = pd.DataFrame({'Name': ['Alice', 'Bob'], 'Age': [25, 30]})
df2 = pd.DataFrame({'Name': ['Charlie', 'David'], 'Age': [35, 40]})
df_concat = pd.concat([df1, df2], axis=0)
print(df_concat)

#17. Merge two DataFrames on a common column.
print("Answer #17.")
df1 = pd.DataFrame({'ID': [1, 2], 'Name': ['Alice', 'Bob']})
df2 = pd.DataFrame({'ID': [1, 2], 'Salary': [50000, 60000]})
df_merged = pd.merge(df1, df2, on='ID')
print(df_merged)

#18. Drop duplicate rows from the DataFrame.
print("Answer #18.")
df_deduped = df.drop_duplicates()
print(df_deduped)

#19. Select multiple columns from the DataFrame.
print("Answer #19.")
selected = df[['Full Name', 'Age']]
print(selected)

#20. Count occurrences of each unique value in a column.
print("Answer #20.")
city_counts = df['City'].value_counts()
print(city_counts)
```

```python
#21. Find the maximum value in a column.
print("Answer #21.")
max_age = df['Age'].max()
print(max_age)

#22. Filter rows based on multiple conditions.
print("Answer #22.")
filtered = df[(df['Age'] > 25) & (df['City'] == 'SF')]
print(filtered)

#23. Calculate the sum of values in a column.
print("Answer #23.")
total_salary = df['Salary'].sum()
print(total_salary)

#24. Apply a function to transform a column.
print("Answer #24.")
df['Age_in_10_years'] = df['Age'].apply(lambda x: x + 10)
print(df)

#25. Sort the DataFrame by multiple columns.
print("Answer #25.")
sorted_multi = df.sort_values(by=['City', 'Age'], ascending=[True, False])
print(sorted_multi)

#26. Add a constant value to a column.
print("Answer #26.")
df['Bonus'] = df['Salary'] + 5000
print(df)

#27. Filter rows containing a specific substring in a column.
print("Answer #27.")
filtered_name = df[df['Full Name'].str.contains('Alice')]
print(filtered_name)

#28. Convert a column to datetime format.
print("Answer #28.")
df['Joining Date'] = pd.to_datetime(['2023-01-01', '2023-02-01', '2023-03-01'])
print(df)

#29. Extract year from a datetime column.
print("Answer #29.")
df['Joining Year'] = df['Joining Date'].dt.year
print(df)

#30. Calculate the correlation between two columns.
print("Answer #30.")
correlation = df[['Age', 'Salary']].corr()
print(correlation)
```

```python
#31. Create a crosstab of two columns.
print("Answer #31.")
crosstab = pd.crosstab(df['City'], df['Joining Year'])
print(crosstab)

#32. Calculate row-wise sum for specific columns.
print("Answer #32.")
df['Total'] = df[['Age', 'Salary']].sum(axis=1)
print(df)

#33. Create a bar plot of a column's values.
print("Answer #33.")
import matplotlib.pyplot as plt
df['Salary'] = df['Salary'].fillna(0)  # Fix chained assignment issue
df['City'].value_counts().plot(kind='bar')
plt.show()

#34. Calculate the standard deviation of a column.
print("Answer #34.")
std_dev = df['Age'].std()
print(std_dev)

#35. Add a prefix to all column names.
print("Answer #35.")
df_prefixed = df.add_prefix('col_')
print(df_prefixed)
```
