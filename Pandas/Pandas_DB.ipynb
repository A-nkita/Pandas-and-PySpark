```python
import time
starttime = time.time()
df = pd.read_csv("Amazon.csv", low_memory= False)

df.to_sql(
    name = "Amazon",
    con = conn,
    if_exists = "fail",
    index = True,
    chunksize = 5000
)

endtime = time.time()
print("Total time = ", endtime-starttime)
```

```python
#Assignment: Fetch data from MySQL, dump it to MSSQL and get email conformation once job is done.

import pandas as pd
import pymysql
from sqlalchemy import create_engine

# 1. Fetch data from MySQL
# MySQL connection details
mysql_conn = pymysql.connect(
    host='localhost',
    user='root',
    password='root',
    db='Pandas_study'
)

# Fetch data
df = pd.read_sql("SELECT * FROM titanic", mysql_conn)
mysql_conn.close()

# 2. Dump data to MSSQL
# MSSQL connection details
server = 'Ankita'
database = 'Pandas_TEST'

mssql_conn = create_engine(f"mssql+pyodbc://@{server}/{database}?trusted_connection=yes&driver=ODBC+Driver+17+for+SQL+Server")

# Write data to MSSQL
df.to_sql('titanic', mssql_conn, if_exists='replace', index=False)

print("Data dumped to MSSQL successfully.")

# 3. Email conformation
```
